## Word Cloud  
Word clouds (also know as text clouds or tag clouds) work in a simple way: the more a specific word appears in a source of textual data (such as a speech, blog post or database), the bigger and bolder it appears in the word cloud.

[word_cloud](https://github.com/amueller/word_cloud/) is a Python package already exists was developed by Andreas_Mueller

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot')
print('Matplotlib version: ', mpl.__version__)
```

```python
import numpy as np #useful for many scientific computing in python
import pandas as pd #primary data structure library
from PIL import Image #converting images into arrays

print('done importing')
```

```python
#import package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS

print('Wordcloud is installed and imported')
```

Let's try to analyze a short novel written by Lewis Carroll titled Alice's Adventures in Wonderland.

```python
#open the file and read it into a variable alice_novel
alice_novel = open(r'C:\Users\khoip\Desktop\alice_novel.txt').read()

print('File downloaded and saved')
```

We use the function set to remove any redundant stopwords.

```python
stopwords = set(STOPWORDS)
```

For simplicity, let's generate a word cloud using only the first 2000 words in the novel

```python
#instantiate a word cloud object
alice_wc = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords
)

#generate the wordcloud
alice_wc.generate(alice_novel)
```

```python
plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

The most common words are Alice, said, little, Queen and so on.  
Let's resize the cloud so that we can see the less frequent words a little better.

```python
fig= plt.figure()
fig.set_figwidth(15)
fig.set_figheight(20)
#show picture after fixed
plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

the word **"Said"** is not informative. Let's add it to our stopwords and re-generate the cloud.

```python
stopwords.add('said')

alice_wc.generate(alice_novel)

fig = plt.figure()
fig.set_figwidth(15)
fig.set_figheight(20)

plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Another cool thing to do is superimposing the words on to a mask of anyshape. Let's try.

```python
alice_mask = np.array(Image.open(r'C:\Users\khoip\Desktop\alice_mask.png'))
australia_mask = np.array(Image.open(r'C:\Users\khoip\Desktop\aus 3.png'))
print('Image loaded and saved!')
```

```python
fig = plt.figure()
fig.set_figwidth(15)
fig.set_figheight(20)

plt.imshow(australia_mask, cmap=plt.cm.gray, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

```python
#Shaping the word cloud according to the mask, using the word cloud package. Using the first 2000 words in the novel.

alice_wc = WordCloud(background_color='white', max_words=2000, mask=australia_mask, stopwords=stopwords)

alice_wc.generate(alice_novel)

fig = plt.figure()
fig.set_figwidth(15)
fig.set_figheight(20)

plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
